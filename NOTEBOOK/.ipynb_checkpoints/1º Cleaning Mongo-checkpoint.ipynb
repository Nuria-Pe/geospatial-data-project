#### I create a DataFrame only with the keys = 'Offices'

```python
b = collection.find({"offices.city": 'Warsaw'}, {"name":1, "offices.city":1, "_id":0, "offices.latitude":1, "offices.longitude":1})
```

```python
pd.DataFrame(b)
```

```python
city = ["city"]
latitud = ["latitude"]
longitud = ["longitude"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
lovemosclaro = []
for element in milista:
    for diccionario in element:
        paralista = {}
        paralista["city"] = getFromDict(diccionario,city)
        paralista["latitude"] = getFromDict(diccionario,latitud)
        paralista["longitude"] = getFromDict(diccionario,longitud)
        lovemosclaro.append(paralista)
```

```python
dataframefinal = pd.DataFrame(lovemosclaro)
dataframefinal.head()
```

```python
dataframefinal.info()
```
