```python
import requests
import json
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
load_dotenv()
```

```python
donde = "warsaw"
def geocode(address):
    """
    Saca las coordenadas de una dirección que le des.
    """
    data = requests.get(f"https://geocode.xyz/{address}?json=1").json()
    try:
        return {
            "type":"Point",
            "coordinates":[float(data["longt"]),float(data["latt"])]}
    except:
        return data
```

```python
data = requests.get(f"https://geocode.xyz/{donde}?json=1").json()
data
```

```python
tok1 = os.getenv("tok1")
tok2 = os.getenv("tok2")
```

```python
enlace = 'https://api.foursquare.com/v2/venues/search'
search = "4bf58dd8d48988d1d3941735"
```

### 1º) STARBUCKS

```python
parametros = { "client_id" : tok1,
          "client_secret" : tok2,
          "v" : "20180323",
          "ll": f"{Varsovia.get('coordinates')[1]},{Varsovia.get('coordinates')[0]}",
          "query": "Starbucks",
}
```

```python
resp = requests.get (url = enlace, params = parametros)
data = json.loads(resp.text)
```

```python
data
```

```python
data.keys()
```

```python
decoding_data = data.get("response")
```

```python
decoding_data.keys()
```

```python
decoding_data
```

```python
from functools import reduce
import operator
```

```python
decoded = decoding_data.get("venues")
```

```python
decoded
```

```python
mapa_nombre = ["name"]
latitud = ["location","lat"]
longitud = ["location","lng"]
```

```python
def getFromDict(diccionario,mapa):
    return reduce(operator.getitem,mapa,diccionario)
```

```python
lovemosclaro = []
for diccionario in decoded:
    paralista = {}
    paralista["name"] = getFromDict(diccionario,mapa_nombre)
    paralista["latitud"] = getFromDict(diccionario,latitud)
    paralista["longitud"] = getFromDict(diccionario,longitud)
    lovemosclaro.append(paralista)
```

```python
dataframefinal = pd.DataFrame(lovemosclaro)
dataframefinal
```

```python
dataframefinal['Type'] = "Starbucks"
```

```python
dataframefinal
```
