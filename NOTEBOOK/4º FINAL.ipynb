```python
import seaborn as sns
import pandas as pd
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
Empresas = pd.read_csv('empresas.csv')
```

```python
Empresas
```

```python
Empresas.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
Starbucks = pd.read_csv('Starbucks.csv')
```

```python
Basketball_Stadium = pd.read_csv('Basketball_Stadium.csv')
```

```python
Music_Venue = pd.read_csv('Music_Venue.csv')
```

```python
queries = [Starbucks, Basketball_Stadium, Music_Venue, Empresas]
```

```python
result = pd.concat(queries)
```

```python
result
```

```python
result.drop(['Unnamed: 0.1', 'city', 'Unnamed: 0'], axis = 1)
```

```python
inicial_lat = 52.2322
inicial_long = 21.0083
```

```python
map_1 = folium.Map(location = [inicial_lat,inicial_long], zoom_start = 15)
map_1
```

```python
for i,row in result.iterrows():
    #popup distrito
    distrito = {
        "location" : [row["latitud"], row["longitud"]]
    }
    
    if row["Type"] == "Starbucks":
        icon = Icon( color = "green",
                    prefix = "fa",
                    icon = "coffee",
                    icon_color = "black"
        )
    elif row["Type"] == "Basketball Stadium":
        icon = Icon( color = "gray",
                    prefix = "fa",
                    icon = "fa-futbol-o",
                    icon_color = "black"
        )
    elif row["Type"] == "Music Venue":
        icon = Icon( color = "orange",
                    prefix = "fa",
                    icon = "glass",
                    icon_color = "black"
        )
    else:
        icon = Icon( color = "red",
                    prefix = "fa",
                    icon = "briefcase",
                    icon_color = "black"
        )
    Marker (**distrito,icon = icon).add_to(map_1)
map_1
```
